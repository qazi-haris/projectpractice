```python
import pandas as pd
Df=pd.read_csv("sampletweets.csv",usecols=['sentiment','text'])
Df=Df.dropna()
Df.loc[Df['sentiment'] == '|negative|','sentiment']=-1
Df.loc[Df['sentiment'] == '|neutral|','sentiment']=0
Df.loc[Df['sentiment'] == '|positive|','sentiment']=1
```

```python
import pandas as pd
Df3=pd.read_csv("train.csv",usecols=['sentiment','text'])
Df3=Df3.dropna()
Df3.loc[Df3['sentiment'] == 'negative','sentiment']=-1
Df3.loc[Df3['sentiment'] == 'neutral','sentiment']=0
Df3.loc[Df3['sentiment'] == 'positive','sentiment']=1
# Df1=Df1[Df1['text']!='Not Available']
Df3=Df3[Df3['sentiment'].isin([1,-1,0])]
```

```python
import chardet
import pandas as pd
import re
def clean_tweet(tweet):
    return ''.join(re.sub(r"(@[A-Za-z0-9]+)|(http\S+)|(#[A-Za-z0-9]+)|(\$[A-Z]+|RT|:)","",tweet))
with open(r'C:\Users\mirza\Desktop\project\AppleTweets.csv', 'rb') as f:
    result = chardet.detect(f.read()) # or readline if the file is large
    df=pd.read_csv(r'C:\Users\mirza\Desktop\project\AppleTweets.csv',encoding=result['encoding'],usecols=['sentiment','text'])
df.loc[df['sentiment'] == '1','sentiment']=-1
df.loc[df['sentiment'] == '3','sentiment']=0
df.loc[df['sentiment'] == '5','sentiment']=1
df=df[df.sentiment!='not_relevant']
```

```python
Df1=pd.concat([Df,Df3,df],ignore_index=True)
Df1
```

```python
import matplotlib.pyplot as plt
neg1 = len(Df1[Df1["sentiment"] == -1])
pos1 = len(Df1[Df1["sentiment"] == 1])
neu1 = len(Df1[Df1["sentiment"] == 0])
# print("negative=\npositive=\nneutral=\nTotal=", neg1,pos1,neu1,neg1+pos1+neu1)
# print('negative:{} \n positive:{} )'.format(negative,positive))
```

```python
def clean_tweet(tweet):
    return ''.join(re.sub(r"(@[A-Za-z0-9]+)|(http\S+)|(#[A-Za-z0-9]+)|(\$[A-Z]+|RT|:)","",tweet))
Df1['cleantweet']=Df1['text'].apply(lambda row: clean_tweet(row))
Df1.cleantweet=Df1.cleantweet.drop_duplicates()
Df1.dropna(inplace=True)
Df1.reset_index(drop=True,inplace=True)
pd.options.display.max_colwidth=1000
```

```python
import matplotlib.pyplot as plt
neg = len(Df1[Df1["sentiment"] == -1])
pos = len(Df1[Df1["sentiment"] == 1])
neu = len(Df1[Df1["sentiment"] == 0])
print(neg,pos,neu,neg+pos+neu)
plt.bar(['negative','positive','neutral'],[neg1,pos1,neu1])
plt.bar(['negative','positive','neutral'],[neg,pos,neu],color='red')
plt.show()
```
