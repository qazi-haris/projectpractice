```python
import pandas as pd
df1=pd.read_csv('AAPL.csv',index_col='Date',parse_dates=True)
df1=df1.drop(columns='Close')
df1.rename(columns={'Adj Close':'Close'},inplace=True)
df0=pd.read_csv('NASDAQ.csv',usecols=['Adj Close','Date'],index_col=['Date'],parse_dates=True)
df0.columns=['NASDAQ']
df2=pd.read_csv('S&P500.csv',usecols=['Adj Close','Date'],index_col=['Date'],parse_dates=True)
df2.columns=['S&P500']
df3=df1.join(df0)
df1=df3.join(df2)
df1.corr(method='pearson')
```

```python
import numpy as np
features=pd.DataFrame(index=df1.index).sort_index()
# Daily Return:
features['f00']=df1.Close
features['f01']=df1.Close/df1.Open-1
features['f02']=df1.Open/df1.Close.shift(-1)-1
features['f03']=df1.Volume.diff()
features['f04']=df1.Volume.diff(50)
features['f05']=df1.Volume.pct_change()
features['f06']=df1.Volume.rolling(5).mean().apply(np.log)
features['f07']=df1.Volume/df1.Volume.rolling(200).mean()-1
features['f08']=df1.Close/df1.Close.ewm(span=50).mean()-1
def z_score(x):
    return (x-x.mean())/x.std()
features['f09']=z_score(df1.Close)
def z_scoreimp(x):
    return (x-x.rolling(window=200,min_periods=20).mean())/x.rolling(window=200,min_periods=20).std()
features['f10']=z_scoreimp(df1.Close)
def rollrank(x):
    return x.rolling(200,min_periods=20).apply(lambda x: pd.Series(x).rank(pct=True)[0])
features['f11']=rollrank(df1.Close)
features['f12']=features['f07'].dropna().rank(pct=True)
features['day']=df1.Close.shift(-5)
features.corr()



```

```python
import sklearn
import ta
features['f13']=ta.momentum.money_flow_index(df1.High,df1.Low,df1.Close,df1.Volume, n=14,fillna=False )
features['rsi']=ta.momentum.rsi(df1.Close,n=14,fillna=False)
features['f14']=features['f13']-features['f13'].rolling(200,min_periods=20).mean()
features['macd']=ta.trend.macd(df1.Close,n_fast=12,n_slow=26,fillna=False)
features['dpo']=ta.trend.dpo(df1.Close,n=20,fillna=False)
features.corr()
```

```python
import numpy as np
features['f15']=features['f05'].apply(np.sign)
for i in features:
    if i=='f03' or 'f04':
        features[i].fillna(features[i].median(),inplace=True)
    features[i].fillna(features[i].mean(),inplace=True)
features['f15']=df1.Volume.diff()
# features['f16']=df1.Open
# features['f17']=df1.High
features['f18']=df1.Volume.diff(1)/df1.shift(-1).Volume
features['f19']=df1.Close.pct_change(5)
features['f20']=(df1.Close-df1.Open)/df1.Open
features['f21']=df1.Close.rolling(window=20).mean()
features['f22']=df1.Close.rolling(window=20).std()
features['f23']=features['f21']+(features['f22']*2)
features['f24']=features['f21']-(features['f22']*2)
features['day']=df1.Close.shift(-30)

features.dropna(inplace=True)

features.corr()




            
    
```
