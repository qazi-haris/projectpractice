```python
import pandas as pd
df=pd.read_csv('labeled.csv',usecols=['sentiment','tweet'])

df.loc[df['sentiment'] < 0 ,'sentiment']=-1
df.loc[df['sentiment'] > 0 ,'sentiment']=1
# df.loc[df['sentiment'] == ,'sentiment']=0
df=df[df['sentiment'].isin([1,-1,0])]
df=df.dropna()
neg1 = len(df[df["sentiment"] == -1])
pos1 = len(df[df["sentiment"] == 1])
neu1 = len(df[df["sentiment"] == 0])
print(neg1,pos1,neu1)
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
# from skleran.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB
from sklearn import svm
import numpy as np
from sklearn.feature_selection import SelectPercentile,f_classif
cv=TfidfVectorizer(sublinear_tf=True,stop_words='english',lowercase=True,ngram_range=(2,2),max_df=0.6)
X=df.tweet
Y=df.sentiment
Y=Y.astype('int')
xtrain=cv.fit_transform(X)

xtraincv,xtestcv,ytraincv,ytestcv=train_test_split(xtrain,Y)
# selector=SelectPercentile(f_classif,percentile=50)
# selector.fit(xtraincv,ytraincv)
# features_trans=selector.transform(xtraincv)
# test_trans=selector.transform(xtestcv)
# clf=svm.SVC(kernel='linear')
# clf.fit(xtraincv,ytraincv)
# print(clf.score(xtestcv,ytestcv))
clf= LogisticRegression()
clf.fit(xtraincv,ytraincv)
print(clf.score(xtestcv,ytestcv))
string=["not bad"]
st=cv.transform(string)
clf.predict(st)
```
